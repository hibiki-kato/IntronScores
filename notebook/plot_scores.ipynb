## 2. Interactive Plots by Species
Run the cells below to generate an interactive plot for each species. You can click on items in the legend to toggle their visibility.

```python
import os
import matplotlib.pyplot as plt
from scripts.plot_scores import plot_eval_scores

# Set matplotlib to output inline interactive plots
plt.ion()
```

### Mouse (Mmus)

```python
sp = "Mmus"
data_dir = f"data/{sp}"

if os.path.exists(data_dir):
    fig = plot_eval_scores(species=sp, data_dir=data_dir, out_png=None, interactive=False)
    plt.show()
else:
    print(f"Skipping {sp}: Data directory not found -> {data_dir}")
```
