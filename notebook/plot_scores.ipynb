## 2. Interactive Plots by Species
Run the cells below to generate an interactive plot for each species using Plotly.

- **Toggle Lines:** Click on a name in the legend on the right to show/hide it.
- **Isolate a Line:** Double-click on a name in the legend to hide all others.
- **Zoom:** Click and drag on the plot to zoom in, double-click to reset.

```python
from scripts.plot_scores import plot_eval_scores_plotly
```
