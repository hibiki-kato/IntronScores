## 2. Generate and View Plots
We can run the Python script `scripts/plot_scores.py` just like the `plot.sh` script does. Then, we display the generated PNG image inline.

```python
import os
from IPython.display import Image, display

# The species you want to plot (Mmus, Athal, Dmel, Hsap)
species_to_plot = ["Mmus", "Athal", "Dmel", "Hsap"]

for sp in species_to_plot:
    print(f"\n{'='*40}")
    print(f"Plotting results for {sp}")
    print(f"{'='*40}")
    
    data_dir = f"data/{sp}/eval_score"
    out_png = f"{sp}_snpr.png"
    
    if os.path.exists(data_dir):
        # Run the plotting script
        !python scripts/plot_scores.py --species {sp} --data_dir {data_dir} --out_png {out_png}
        
        # Display the result
        if os.path.exists(out_png):
            display(Image(filename=out_png))
    else:
        print(f"Skipping {sp}: Data directory not found -> {data_dir}")
```
