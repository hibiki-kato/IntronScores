## 2. Generate and View Plots
We import our `plot_eval_scores` function directly from the `scripts` package, and then display the interactive plots.

```python
import os
import matplotlib.pyplot as plt
from scripts.plot_scores import plot_eval_scores

# Set matplotlib to output inline interactive plots
plt.ion()

# The species you want to plot (Mmus, Athal, Dmel, Hsap)
species_to_plot = ["Mmus", "Athal", "Dmel", "Hsap"]

for sp in species_to_plot:
    print(f"\n{'='*40}")
    print(f"Plotting results for {sp}")
    print(f"{'='*40}")
    
    data_dir = f"data/{sp}"
    
    if os.path.exists(data_dir):
        # Generate the plot interactively, without saving a PNG
        fig = plot_eval_scores(species=sp, data_dir=data_dir, out_png=None, interactive=False)
        
        # Force the plot to display now
        plt.show()
    else:
        print(f"Skipping {sp}: Data directory not found -> {data_dir}")
```
